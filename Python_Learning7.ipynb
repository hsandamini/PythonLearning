# 🔁 Loops in Python

Loops allow you to automate repetitive tasks by running a block of code multiple times.

1. **Definite Iteration (`for` loops):** Executed a predictable number of times based on an iterable sequence.
2. **Indefinite Iteration (`while` loops):** Controlled purely by a truth condition; runs continuously until that condition breaks.
3. **Control Signals (`break`, `continue`, `pass`):** Keywords that override standard operational sequences mid-cycle.
4. **The `else` Clause:** A unique Python structural rule that runs code *only* if a loop finishes cleanly without encountering a `break`.

### 1. Basic for Loop (Counting)

```python
# Basic for loop using range to count from 1 to 5
for i in range(1, 6):
    print("Counting:", i)
```

### 2. Basic while Loop (State Condition)

```python
# Basic while loop that runs as long as a counter condition is met
count = 1

while count <= 5:
    print("While count is:", count)
    count += 1  # Updates the state to eventually end the loop
```

### 3. Basic Collection Loop

```python
# Iterating directly through items in a list sequence
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("Fruit name:", fruit)
```

### 4. Looping Through a Collection

```python
fruits = ["lemon", "cherry", "banana"]

for x in fruits:
    print(x)
```

### 5. Looping Through a String

```python
# Looping through a string
for i in "Python":
    print(i)
```

### 6. The continue Command (Skipping Steps)

```python
# Skipping a specific step mid-cycle without stopping the entire loop
for i in range(1, 6):
    if i == 3:
        print("-> Found 3: Skipping this step using 'continue'")
        continue  # Skips print(i) for number 3 and jumps right to number 4
    print("Number:", i)
```

### 7. The break Command (Instant Shutdown)

```python
# Aborting a loop early as soon as a condition is satisfied
for i in range(1, 6):
    if i == 4:
        print("-> Target 4 reached: Killing the loop using 'break'")
        break  # Shuts down the entire loop immediately
    print("Processing item:", i)
```

### 8. The pass Command (The Structural Placeholder)

```python
# pass is a null placeholder. It prevents a syntax error while keeping a loop empty.
for i in range(1, 4):
    if i == 2:
        pass  # Do nothing here right now (tells Python to keep moving)
    print("Current loop item:", i)
```

### 9. Nested Matrix Loop

```python
# A nested loop structure where an inner loop cycles fully for every single outer pass
# Great for searching multi-dimensional data grids or coordinate spaces
for row in ["Row A", "Row B"]:
    print(f"--- Starting {row} ---")
    for column in range(1, 4):
        print(f"Checking coordinate: [{row}, Col {column}]")
```

### 10. for ... else Loops

```python
# The 'else' block attached to a loop only fires if the loop finishes WITHOUT hitting a 'break'.
# This functions like a built-in search success/failure indicator.

search_target = "David"
employees = ["Alex", "Emma", "Sarah", "Sophia"]

print(f"Initiating security search for: {search_target}...")

for person in employees:
    print(f"Scanning profile: {person}")
    if person == search_target:
        print(f"Success! Found match for {search_target}. Access granted.")
        break  # If found, loop breaks, and the 'else' block below is SKIPPED
else:
    # This runs ONLY if the loop finishes scrolling through the whole list without a break
    print(f"Search complete: '{search_target}' is not in the database directory.")
```

### 🔢 The `range(start, stop, step)` Parameter Matrix

The `range()` function can take up to three arguments to finely customize a sequence of numbers:
* **Start (`2`):** The first number where the sequence begins.
* **Stop (`30`):** The upper boundary limit. The loop stops *before* reaching this number.
* **Step (`3`):** The increment value. Instead of counting by 1, Python adds 3 on each iteration.

```python
for x in range(2, 30, 3):
    print(x)
```

### 🔄 Advanced Step Counting Patterns
* **Positive Step:** Increments forward (e.g., counting by 2s or 10s).
* **Negative Step:** Counts backward to create a decrementing loop or countdown. 
* **Boundary Rule:** The loop always stops exactly one step *before* hitting the specified `stop` limit.

### 1. Counting by Even Numbers (Step = 2)

```python
# Starts at 2, ends before 11, increments by 2
for x in range(2, 11, 2):
    print("Even number:", x)
```

### 2. Counting backward (Negative Step)

```python
# Countdown: Starts at 5, ends before 0, decrements by -1
for x in range(5, 0, -1):
    print("Countdown:", x)
print("Blast off!")
```

### 3. Multiplying inside the Step sequence

```python
# Starts at 10, ends before 51, increments by 10
for x in range(10, 51, 10):
    print(f"Interval checkpoint: {x}")
```

```python
for num in range(-2,-5,-1):
    print(num, end=", ")
```

## 💡 Essential Python Logic Rules 

### 1. Truthy vs Falsy Behavior (`if -3`)
When you pass a raw number directly into a conditional statement like `if -3:`, Python automatically runs a data evaluation check:
* **The Falsy Group:** Only numeric values that are **exactly zero** (`0`, `0.0`, `0j`) or empty containers (like an empty string `""` or list `[]`) evaluate to `False`.
* **The Truthy Group:** **Any non-zero value** is considered `True`. Because negative numbers (like `-3`) are not zero, they always evaluate to **True**.

### 2. The `for...else` Structural Clause
The `else` block attached to a loop has a very specific behavioral rule:
* It executes **only when the loop terminates naturally** (after successfully cycling through every item in a sequence or range).
* If the loop terminates abruptly because it hit a **`break`** statement, the `else` block is completely skipped.


```python
if -3:
    print("This runs because -3 is a non-zero number, making it TRUTHY!")
else:
    print("This will never print.")

# Checking it directly using the built-in bool function
print("bool(-3) evaluates to:", bool(-3))
```

## 📐 Nested Loop Multiplication Rule 

When dealing with a nested loop configuration where code is incremented inside the deepest layer, we can determine the total execution count using simple multiplication:

$$\text{Total Increments} = (\text{Outer Loop Iterations}) \times (\text{Inner Loop Iterations})$$

### 🔬 Breakdown of the Puzzle:
1. **Outer Loop Sequence:** `range(10)` runs from `0` to `9`, yielding exactly **10 cycles**.
2. **Inner Loop Sequence:** `range(-1, -10, -1)` counts backwards (`-1, -2, -3, -4, -5, -6, -7, -8, -9`), yielding exactly **9 cycles**.
3. **Execution Total:** The inner statement `x += 1` triggers $10 \times 9 = 90$ total times.

```python
x = 0
for i in range(10):
    for j in range(-1, -10, -1):
        x += 1

print("The final value of x is:", x)
```
