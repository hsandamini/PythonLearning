# Python Identifiers

### 🚫 Rules for Naming Python Variables

* We **cannot** use Python **keywords** as variable names (identifiers).
* We **cannot** use special symbols like `@`, `#`, `$`, or `%` anywhere in a variable name.
* We **cannot** start a variable name with a number.
* However, we **can** use numbers anywhere else in the name (middle or end).
* The only special character allowed is the underscore (`_`).


```python
# WRONG: Using a reserved keyword causes a SyntaxError
False = 100
```

### Rule 1: Keywords

```python
# CORRECT: Modify the name slightly so it is no longer a keyword
is_false = 100
```

```python
is_false
```

### Rule 2: Special Symbols

```python
# WRONG: Using special characters like @, #, $, % causes an error
@name = "Tom"
user$money = 500
```

```python
# CORRECT: Use regular text characters or numbers instead
name = "Tom"
user_money = 500
```

```python
print(name, user_money)
```

### Rule 3: Starting with a number

```python
# WRONG: Starting the variable name with a digit causes an error
1variable = "name"
```

```python
# CORRECT: Start with a letter, and put numbers at the middle or end
variable1 = "name"
```

```python
variable1
```

### Rule 4: Underscores

```python
#  CORRECT: The underscore is the only special character allowed!
_system_status = "Active"
total_price = 19.99

print(_system_status)
print(total_price)
```

# ⚙️ Python Statements

A **statement** is an instruction that the Python interpreter can execute. 
Python supports single-line statements, multi-line statements (using `\` or brackets), and multiple statements on one line separated by a semicolon `;`.


### 1. Single-line statement

```python
# A simple assignment statement
message = "Hello, Python!"
print(message)
```

### 2. Multi-line statement

#### Option 1: Using a backslash\

```python
# The backslash tells Python the statement continues on the next line
total_sum = 10 + 20 + \
            30 + 40 + \
            50

print(total_sum)
```

#### Option 2: Using parentheses

```python
# Anything inside parentheses, brackets [], or braces {} can span multiple lines automatically
total_elements = (10 + 20 +
                  30 + 40 +
                  50)

print(total_elements)
```

```python
# Multi-line statement using a List []
fruits = [
    "banana",
    "apple",
    "orange"
]

print(fruits)
```

#### Option 3: Multiple statements in one line

```python
# Two separate statements on one line
x = 5; y = 10
print(x + y)
```

# 📐 Python Indentation

In Python, indentation (spaces at the start of a line) is not just for readability. It is a strict syntax requirement used to define blocks of code (like loops, functions, and if-statements). 

* The standard convention is to use **4 spaces** or **1 Tab** per indentation level.
* Incorrect indentation will throw an `IndentationError`.
* Every time a line ends with a colon (`:`), the next line **must** step inward by 4 spaces (or 1 Tab).
* In nested blocks (like an `if` inside a `for` loop), the inner code steps inward twice (8 spaces).

### Correct example

```python
# CORRECT: Proper nested indentation levels
for i in range(1, 10):
    print(i)          # Indented once (inside the loop)
    if i == 6:        # Indented once (inside the loop)
        break         # Indented twice (inside the 'if' block)
```

### Wrong example

```python
# WRONG: print(i) must be pushed inward under the loop
for i in range(1, 10):
print(i)
    if i == 6:
        break
```

# 💬 Python Comments

Comments are non-executable lines used to document code behavior. They help make scripts self-explanatory for developers while being completely ignored by the Python interpreter.

* **Single-line comments** start with a hash symbol (`#`).
* **Multi-line notes/docstrings** use triple quotes (`"""` or `'''`).

### 1. Single-line comments

```python
# This is a full single-line comment explaining the variable below
user_age = 25  

print(user_age)  # This is an inline comment explaining what is being printed
```

### 2. Multi-line comments

```python
"""
This is a multi-line note block.
We can write as many lines as you want here.
Python will look at this text and safely skip over it.
"""
username = "Alex"
print(username)
```

### 3. Using comments to disable codes

```python
print("Step 1: Database connected.")
# print("Step 2: Sending diagnostic emails...") # This line is temporarily disabled and won't run
print("Step 3: App initialized successfully.")
```
