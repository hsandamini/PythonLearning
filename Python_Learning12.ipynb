# 🔨 Functions in Python

### 🔨 1. Defining Functions
To create a function, write the **`def`** keyword followed by a unique function name and parentheses `()`.
* **The Colon (`:`):** Declares the start of the code block.
* **Indentation:** The action code underneath **must** be indented by 4 spaces.
* **Execution:** Defining a function does not run it. You must explicitly **call** the function using its name with parentheses.

```python
# 1. Defining the function structure
def greet_user():
    print("Hello! Welcome to Python functions.")

# 2. Calling the function to execute it
greet_user()
```

### 📦 2. Function Arguments and Parameters
Functions can accept input variables to process dynamic data layouts:
* **Parameter:** The variable listed inside the function definition parentheses (the placeholder).
* **Argument:** The actual concrete value sent to the function when it is called.

```python
# 'name' is the input parameter (the placeholder variable)
def welcome_member(name):
    print(f"Welcome to the portal, {name}!")

# Passing unique string arguments into the function call
welcome_member("Alice")
welcome_member("Bob")
```

### 📋 3. Inserting a Collection as an Argument
You can send any data type into a function as an argument, including mutable collection objects like **Lists**. The function will receive and loop through the collection structure seamlessly.

```python
# Defining a function that processes data sequences
def loop_food_basket(food_list):
    for item in food_list:
        print(f"Processing basket item: {item}")

# Creating a standard list collection
myfruits = ["fig", "apple", "banana"]

# Sending the entire list container as a single argument
loop_food_basket(myfruits)
```

### 📤 4. Return and the Expression
To let a function send a calculated data result back to the main script runner, use the **`return`** statement. 
* **Crucial Rule:** `return` exits the function immediately. Any code written beneath it inside the function block will be completely ignored!

```python
# Defining a function that calculates a math operation
def multiply_by_five(number):
    return number * 5

# Saving the returned output value into a variable
result = multiply_by_five(12)
print("The calculated return value is:", result)
```
