# 🔢 Numbers in Python

Python supports three built-in numerical data types:
1. **`int` (Integer):** Whole numbers of unlimited length (e.g., `10`, `-500`).
2. **`float` (Floating Point):** Numbers containing decimals or scientific notation (e.g., `10.5`, `3.14`).
3. **`complex` (Complex Numbers):** Numbers with a real and an imaginary part, written with a `j` suffix (e.g., `2 + 3j`).

### 1. Working with Integers (int)

```python
# Standard whole numbers
positive_int = 42
negative_int = -1050
huge_int = 12345678901234567890

print(positive_int, type(positive_int))
print(negative_int, type(negative_int))
print(huge_int, type(huge_int))
```

### 2. Working with Floating-Point Numbers (float)

```python
# Decimals and scientific E-notation
pi_val = 3.14159
negative_decimal = -0.5

# 2e4 means 2 * 10^4 = 20000.0
scientific_num = 2e4  

print(pi_val, type(pi_val))
print(negative_decimal, type(negative_decimal))
print(scientific_num, type(scientific_num))
```

### 3. Working with Complex Numbers (complex)

```python
# Real + Imaginary parts
z1 = 4 + 7j
z2 = -3j  # Purely imaginary

print(z1, type(z1))
print("Real part:", z1.real)
print("Imaginary part:", z1.imag)
```

# 📋 Lists in Python

A **List** is an ordered, changeable collection of items enclosed in square brackets `[]`.

### 🔑 Key Characteristics of Lists:
1. **Ordered:** Items maintain their specific insertion sequence. Each item has a fixed index starting at `0`.
2. **Mutable:** You can freely add, remove, or modify items after the list is created.
3. **Allows Duplicates:** The same value can appear multiple times at different positions.
4. **Heterogeneous:** A single list can store a mix of completely different data types.

### 1. Creating and Printing Lists

```python
# Creating lists with single and mixed data types
hardware = ["CPU", "GPU", "RAM", "SSD"]
mixed_list = ["Python", 2026, 99.9, True]

print("Hardware List:", hardware)
print("Mixed Data List:", mixed_list)
```

### 2. Accessing Items Using Indexing

```python
fruits = ["fig", "lemon", "banana", "apple"]

print("First item (index 0):", fruits[0])
print("Third item (index 2):", fruits[2])
print("Last item (index -1):", fruits[-1])
```

### 3. Proving Mutability (Modifying Items)

```python
prices = [10, 20, 30, 40]
print("Original prices:", prices)

# Change the item at index 1 from 20 to 99
prices[1] = 99
print("Modified prices:", prices)
```

### 4. Common List Methods (append and len)

```python
todo_list = ["Code", "Read"]

# Add a new item to the very end of the list
todo_list.append("Exercise")
print("Updated list:", todo_list)

# Find out how many items are in the list total
print("Total number of items:", len(todo_list))
```

# 🔒 Tuples in Python

A **Tuple** is an ordered, unchangeable collection of items enclosed in parentheses `()`.

### 🔑 Key Characteristics of Tuples:
1. **Ordered:** Items maintain a strict sequence, accessible by zero-based indexing.
2. **Immutable:** Once defined, you cannot append, alter, or delete individual items. This makes them faster and safer than lists.
3. **Allows Duplicates:** The exact same value can be stored multiple times.
4. **Heterogeneous:** Can contain any mixture of integers, strings, booleans, or floats.

### 1. Creating and Printing Tuples

```python
# Creating tuples using standard parentheses
dimensions = (1920, 1080)
mixed_tuple = ("Success", 200, True, 4.5)

print("Dimensions Tuple:", dimensions, type(dimensions))
print("Mixed Data Tuple:", mixed_tuple, type(mixed_tuple))
```

### 2. Accessing Items and Finding Length

```python
letters = ('A', 'B', 'C', 'D')

print("First item (index 0):", letters[0])
print("Last item (index -1):", letters[-1])
print("Total number of items:", len(letters))
```

### 3. Proving Immutability (The Error Log)

```python
# TESTING IMMUTABILITY: This cell will intentionally throw a TypeError
numbers = (1, 2, 3, 4)
numbers[0] = 99
```

### 4. The Single Item Tuple Catch

```python
# This is interpreted as a regular String
wrong_tuple = ("banana")
print("Without comma type:", type(wrong_tuple))

# This is recognized correctly as a Tuple
correct_tuple = ("banana",)
print("With comma type:", type(correct_tuple))
```

# 🔤 Strings in Python

A **String** is a text-based data type enclosed in either single quotes (`'`) or double quotes (`"`).

### 🔑 Key Characteristics of Strings:
1. **Ordered:** Every character in a string has a specific position (index) starting at `0`.
2. **Immutable:** Once a string is created, you cannot modify its characters individually. 
3. **Multi-line Support:** You can create long blocks of text using triple quotes (`"""` or `'''`).

### 1. Creating and Printing Strings

```python
# Strings can use single, double, or triple quotes
course_name = "Python Data Science Bootcamp"
greeting = 'Hello, World!'

print(course_name, type(course_name))
print(greeting, type(greeting))
```

### 2. String Indexing (Accessing Characters)

```python
word = "Python"

print("First character (index 0):", word[0])
print("Second character (index 1):", word[1])
print("Last character (index -1):", word[-1])
```

### 3. Proving Immutability (The Error Log)

```python
# TESTING IMMUTABILITY: This cell will intentionally crash if run
sample_text = "Jello"
sample_text[0] = "H"
```

### 4. String Concatenation and Repetition

```python
first = "Data"
second = "Science"

# Join strings with a space in between
full_topic = first + " " + second
print("Concatenation:", full_topic)

# Repeat a string multiple times
alert = "Help! " * 3
print("Repetition:", alert)
```

### 5. Finding String Length

```python
phrase = "Learn Python"
print("Total characters:", len(phrase))
```

# 🌀 Sets in Python

A **Set** is an unordered collection of unique items enclosed in curly braces `{}`.

### 🔑 Key Characteristics of Sets:
1. **Unordered:** Items do not have a fixed index or sequence. Every time you print a set, the items may appear in a different order.
2. **Unindexed:** Because they are unordered, you cannot access items using numbers like `my_set[0]`.
3. **Unique Elements Only:** If you try to add duplicate items, Python automatically deletes the duplicates.
4. **Mutable:** You can add or remove items from a set, but the items inside the set must be of an unchangeable (immutable) data type, like strings, integers, or tuples.

### 1. Creating a Set and Automatic Duplicate Removal

```python
# Duplicate numbers inside the curly braces
id_numbers = {10, 20, 20, 30, 40, 40, 50}

# Python automatically removes duplicates in the output block
print("Set Output:", id_numbers)
print("Data Type:", type(id_numbers))
```

### 2. Proving Sets are Unindexed (The Error Log)

```python
# TESTING INDEXING: This cell will intentionally fail if run
colors = {"red", "green", "blue"}

# Sets do not support indexing! 
print(colors[0])
```

### 3. Adding and Removing Items

```python
skills = {"Python", "SQL"}
print("Original set:", skills)

# Add a single item
skills.add("Git")
print("After adding:", skills)

# Remove an item
skills.remove("SQL")
print("After removing:", skills)
```

### 4. The Empty Set Trap

```python
# This creates an empty dictionary, NOT an empty set
wrong_empty = {}
print("Type of {}:", type(wrong_empty))

# This is the correct way to initialize an empty set
correct_empty = set()
print("Type of set():", type(correct_empty))
```

# 📖 Dictionaries in Python

A **Dictionary** is an ordered collection of key-value pairs enclosed in curly braces `{key: value}`.

### 🔑 Key Characteristics of Dictionaries:
1. **Key-Value Structure:** Every data entry consists of a unique identifier (Key) mapped to a data object (Value).
2. **Ordered:** Since Python 3.7, dictionaries maintain the exact order in which items were inserted.
3. **Mutable:** You can change, add, or remove items freely after the dictionary is created.
4. **Unique Keys:** Duplicate keys are not allowed. If you try to assign a value to an existing key, the old value gets overwritten.

### 1. Creating and Printing a Dictionary

```python
# Storing a user profile record using clean key-value pairs
employee = {
    "name": "David",
    "role": "Data Analyst",
    "experience_years": 4,
    "skills": ["Python", "SQL", "Tableau"]
}

print(employee)
print(type(employee))
```

### 2. Accessing Values Using Their Keys

```python
user_profile = {"username": "cyber_coder", "status": "Active", "score": 95}

# Pass the specific key name inside square brackets []
print("Username:", user_profile["username"])
print("User Score:", user_profile["score"])
```

### 3. Modifying Values and Overwriting Duplicate Keys

```python
car_info = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print("Original:", car_info)

# 1. Modifying an existing value
car_info["year"] = 2024

# 2. Overwriting happens if you supply the same key again
car_info["model"] = "Mach-E" 

print("Updated:", car_info)
```

### 4. Adding New Items and Deleting Existing Ones

```python
student = {"name": "Emma", "grade": "A"}

# Add a completely new key-value pair on the fly
student["id"] = 10452
print("After adding ID:", student)

# Remove a pair using the 'del' statement
del student["grade"]
print("After deleting grade:", student)
```
