# 📖 Python Dictionaries: Definition & Core Features

A **Dictionary** is a collection of data elements stored as key-value associations wrapped inside curly braces `{key: value}`.

### 🔑 Key Architectural Features of a Dictionary:
1. **Key-Value Structure:** Every individual data item consists of a lookup label (Key) mapped directly to a data storage target (Value).
2. **Keys Must Be Unique:** Duplicate keys are strictly forbidden. If you try to assign a value to an existing key, the old value gets completely overwritten.
3. **Keys Must Be Immutable:** A dictionary key must be an unchangeable data type, such as strings, integers, or tuples. Values, however, can be absolutely anything (lists, numbers, strings, or even other dictionaries).
4. **Ordered Sequence:** Since Python 3.7, dictionaries maintain a strict insertion sequence. Items stay in the exact order you defined them.
5. **Mutable Container:** You can dynamically append new key-value pairs, adjust existing records, or clear information fields.

### 🏷️ 1. Instantiating Dictionaries and Fetching Records
Dictionaries are initialized using curly braces containing `key: value` declarations separated by commas. To read a specific value, you pass its matching key label inside square brackets `[]`.

```python
# Initializing a structured data dictionary mapping user attributes
mydict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

print("Dictionary Object:", mydict)
print("Data Container Type:", type(mydict))

# Fetching a single record value using its associated key label string
car_brand = mydict["brand"]
print("The accessed brand value is:", car_brand)
```

### 📏 2. Counting Total Attributes via `len()`
When applied to a dictionary, the built-in **`len()`** function counts the total number of independent **key-value pairs** currently tracked inside the container.

```python
mydict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

# Counts the keys ("brand", "model", "year") -> 3 pairs total
print("Total key-value pairings inside dictionary:", len(mydict))
```

### ⚠️ 3. Key Value Overwriting Constraints and Mixed Datatypes
While dictionary keys are constrained to unique immutable entities, the matching values can consist of heterogeneous combinations (integers, strings, booleans, or lists). Providing a duplicate key will overwrite the previous entry.

```python
# 1. Instantiating a mixed-datatype layout with a duplicate key error
car_profile = {
    "brand": "Ford",
    "electric": False,
    "year": 1964,
    "colors": ["red", "white", "blue"],
    "year": 2026  # ⚠️ Duplicate key! This permanently overwrites 1964
}

# Notice that "year" updates to 2026, and the old 1964 value is completely lost
print("Updated Dictionary Layout:", car_profile)
```

### 🏗️ 4. Dictionary Construction via `dict()`
You can also instantiate a dictionary by utilizing Python's built-in **`dict()`** constructor function. When using this construction format, keys are passed as named keyword parameters *without* quotation marks, using a standard assignment operator (`=`).

```python
# Initializing a dictionary through the explicit dict() function wrapper
sample_dict = dict(name="John", age=36, country="Norway")

print(sample_dict)
print(type(sample_dict))
```

### 🔍 5. Reading Values Safely Using the `.get()` Method
While square brackets `mydict["key"]` work for accessing values, if a key does not exist, your program will crash with a `KeyError`. The **`.get(key)`** method retrieves values safely. If the key is missing, it returns `None` instead of crashing.

```python
# Initializing a fresh dictionary matching the image context
newcars = {
    "brand": "BMW",
    "model": "M4",
    "year": 2026
}

# Fetching the value associated with the "brand" key
x = newcars.get("brand")
print(x)
```

### 🏷️ 6. Extracting Lookup Labels Using the `.keys()` Method
The **`.keys()`** method extracts a dynamic view object containing all the lookup keys defined inside the dictionary. This object automatically updates if you append new items to the dictionary later.

```python
# Extracting all active key labels from the newcars dictionary container
x = newcars.keys()
print(x)
```

### ➕ 7. Appending New Fields to a Dictionary on the Fly
Because dictionaries are **mutable**, you can dynamically insert a brand-new field at any time. You use the bracket assignment syntax: `mydict["new_key"] = value`. If the key label does not exist inside the dictionary, Python appends it automatically to the very end of the collection.

```python
# Initializing the base dictionary context to match your screenshot
newcars = {
    "brand": "BMW",
    "model": "X5",
    "year": 2017
}

# Dynamically appending a brand-new key-value pair to the record
newcars["color"] = "black"

# Print the updated dictionary container to verify the modification
print(newcars)
```

### 📊 8. Extracting Storage Targets Using the `.values()` Method
The **`.values()`** method extracts a dynamic view object containing all the actual stored data values inside the dictionary, ignoring the key lookup labels completely. Like `.keys()`, this view updates automatically if the underlying data values are modified or if new entries are appended later.

```python
# Initializing the base dictionary container from Chapter 8
newcars = {
    "brand": "BMW",
    "model": "X5",
    "year": 2017,
    "color": "black"
}

# Extracting all active data storage values from the dictionary
x = newcars.values()
print(x)
```

### 🔄 9. Modifying an Existing Record Value
If you use the bracket assignment syntax `mydict["existing_key"] = new_value` on a key label that already exists inside the dictionary, Python will locate that key and overwrite its old value with the new one.

```python
# Initializing the base dictionary container from Chapter 9
newcars = {
    "brand": "BMW",
    "model": "X5",
    "year": 2017,
    "color": "black"
}

# Overwriting the "year" key value from 2017 to 2019
newcars["year"] = 2019
print(newcars)
```

### 📦 10. Extracting Key-Value Tuples Using the `.items()` Method
The **`.items()`** method returns a dynamic view object containing a list of matching **tuples**. Each tuple contains exactly two elements: the key at index `0` and its mapped value at index `1`. This is the standard method used when looping through a dictionary.

```python
# Extracting all key-value pairings as an iterable list of tuples
x = newcars.items()
print(x)
```

### 🔍 11. Checking Key Existence Using the `in` Operator
You can check if a specific lookup label exists inside a dictionary using the **`in`** membership operator. 

⚠️ **Crucial Behavior Note:** The `in` operator **only searches the Keys**, not the actual data storage values.

```python
# Verifying if the key label "color" is actively tracked inside newcars
if "color" in newcars:
    print("Yes, color is in newcars!")
```

### 🔄 12. Modifying Records and Merging via the `.update()` Method
The **`.update(iterable)`** method updates a dictionary with elements from another dictionary object or from an iterable of key-value pairs. If the key label already exists inside your dictionary, its value is overwritten. If the key is new, it is appended to the end of the collection.

```python
# Initializing the base dictionary container from Chapter 12
newcars = {
    "brand": "BMW",
    "model": "X5",
    "year": 2019,
    "color": "black"
}

# Modifying the "color" key value from "black" to "white" using update()
newcars.update({"color": "white"})
print(newcars)
```

### 🗑️ 13. Extracting and Deleting Specific Fields using `.pop()`
The **`.pop(key)`** method removes a specified key from the dictionary and returns its corresponding value. If you print the dictionary afterward, that key-value pair is completely gone. 

⚠️ **Crucial Warning:** Unlike list popping, you **must** pass a specific key argument into a dictionary's `.pop()` method. Leaving it completely blank will throw a `TypeError`.

```python
# Current dictionary state: {'brand': 'BMW', 'model': 'X5', 'year': 2019, 'color': 'white'}

# Remove the "year" field completely from the record
newcars.pop("year")
print(newcars)
```

### 🎲 14. Removing the Last Inserted Pair using `.popitem()`
The **`.popitem()`** method removes and returns the last key-value pair inserted into the dictionary. The returned item is formatted as a **tuple** containing `(key, value)`. This method takes no arguments. Calling it on an empty dictionary will cause a `KeyError`.

```python
# Initializing the base dictionary container from Chapter 14
newcars = {
    "brand": "BMW",
    "model": "X5",
    "color": "white"
}

# Remove and return the last inserted item
popped_pair = newcars.popitem()
print("Returned tuple: ", popped_pair)

# Print the remaining dictionary container to verify the deletion
print(newcars)
```

### 🗑️ 15. Removing a Specific Key using the `del` Keyword
The **`del`** keyword can be used with a specific key label like `del mydict["key"]` to permanently delete that entry from the dictionary container. If the key is not found, Python throws a standard `KeyError`.

```python
# Initializing the dictionary container context from Chapter 15
newcars = {
    "brand": "BMW",
    "model": "X5"
}

# Permanently delete the "brand" key and its corresponding value
del newcars["brand"]
print(newcars)
```

### 🧹 16. Emptying a Dictionary Container using `.clear()`
The **`.clear()`** method strips out every single key-value pair from a dictionary at once. Unlike deleting the entire variable from memory, this method keeps the container alive as a completely blank dictionary literal object (**`{}`**).

```python
# Current dictionary state from previous step: {'model': 'X5'}

# Clear all remaining elements from the dictionary container completely
newcars.clear()
print(newcars)
```
