# 🧮 Python Arithmetic Operators

Arithmetic operators are used with numeric values to perform common mathematical calculations:

| Operator | Name | Example | Explanation |
| :---: | :--- | :---: | :--- |
| `+` | Addition | `x + y` | Adds values together |
| `-` | Subtraction | `x - y` | Subtracts one value from another |
| `*` | Multiplication | `x * y` | Multiplies values |
| `/` | Division | `x / y` | Divides values (always returns a **float**) |
| `%` | Modulus | `x % y` | Returns the **remainder** of a division |
| `**` | Exponentiation | `x ** y` | Raises a number to a power ($x^y$) |
| `//` | Floor Division | `x // y` | Divides and rounds down to the nearest whole number |

### 1. Basic Math Operations (+, -, *, /)

```python
x = 15
y = 4

print("Addition (15 + 4):", x + y)
print("Subtraction (15 - 4):", x - y)
print("Multiplication (15 * 4):", x * y)
print("Division (15 / 4):", x / y)  # Notice this outputs a decimal float 3.75
```

### 2. Modulus Operator (% - Remainder)

```python
# Modulus returns whatever is left over after division
print("Remainder of 15 divided by 4 (15 % 4):", 15 % 4) 

# Example: Checking if a number is even (remainder will be 0)
print("Is 10 even? (10 % 2):", 10 % 2)
```

### 3. Exponentiation (** - Power)

```python
# Calculating 2 to the power of 3 (2 * 2 * 2)
base = 2
exponent = 3

print("2 cubed (2 ** 3):", base ** exponent)
```

### 4. Floor Division (// - Integer Division)

```python
# Standard division vs Floor division
print("Standard Division (15 / 4):", 15 / 4)
print("Floor Division (15 // 4):", 15 // 4)  # Drops the .75 entirely
```

### 📐 Operator Precedence (BODMAS / PEMDAS)

When a single line of code combines multiple arithmetic operators, Python relies on a strict order of operations (**Operator Precedence**) to decide which math rule to execute first.

In Python's precedence hierarchy:
1. **Exponentiation (`**`)** has higher priority.
2. **Multiplication (`*`)** has lower priority.

If operators have the exact same priority level (like the two multiplication signs here), Python reads and executes the code sequentially from **left to right**.

```python
result = 2 * 3 ** 3 * 4
print("The output of print(2 * 3 ** 3 * 4) is:", result)
```

# ⚖️ Python Comparison Operators

Comparison operators assess the relationship between two operands and return a Boolean value (`True` or `False`).

| Operator | Name | Example | Explanation |
| :---: | :--- | :---: | :--- |
| `==` | Equal to | `x == y` | True if values are identical |
| `!=` | Not equal to | `x != y` | True if values are completely different |
| `>` | Greater than | `x > y` | True if left value is strictly larger |
| `<` | Less than | `x < y` | True if left value is strictly smaller |
| `>=` | Greater than or equal to | `x >= y` | True if left value is larger or identical |
| `<=` | Less than or equal to | `x <= y` | True if left value is smaller or identical |

### 1. Equality & Inequality (==, !=)

```python
a = 10
b = 20

# Checking if values match or differ
print("Is 'a' equal to 'b' (10 == 20)?", a == b)
print("Is 'a' not equal to 'b' (10 != 20)?", a != b)
```

### 2. Greater Than & Less Than (>, <, >=, <=)

```python
x = 5
y = 5
z = 12

print("Is z greater than x (12 > 5)?", z > x)
print("Is x less than z (5 < 12)?", x < z)

# Checking boundary inclusion with equal-to conditions
print("Is x greater than or equal to y (5 >= 5)?", x >= y)
print("Is z less than or equal to y (12 <= 5)?", z <= y)
```

### 3. String Comparisons

```python
word1 = "apple"
word2 = "banana"

print("Is 'apple' less than 'banana' alphabetically?", word1 < word2)
print("Is 'apple' equal to 'Apple' (case-sensitive)?", "apple" == "Apple")
```

# 🧠 Python Logical Operators

Logical operators allow you to evaluate multiple conditions together to form complex logical expressions.

| Operator | Description | Rule |
| :---: | :--- | :--- |
| `and` | Logical AND | Returns `True` **only if both** statements are true. |
| `or` | Logical OR | Returns `True` if **at least one** of the statements is true. |
| `not` | Logical NOT | Reverses the result (returns `False` if the result is true). |

### 1. The and Operator (Both must be True)

```python
age = 25
has_license = True

# Both conditions must evaluate to True for the overall expression to be True
can_drive = (age >= 18) and (has_license == True)

print("Can the user drive?", can_drive)
```

### 2. The or Operator (At least one must be True)

```python
is_weekend = False
is_holiday = True

# Only one condition needs to be True to grant a day off
stay_home = is_weekend or is_holiday

print("Can I stay home today?", stay_home)
```

### 3. The not Operator (Reverse the Logic)

```python
is_game_over = False

# 'not' turns False into True, and True into False
print("Is the game still running?", not is_game_over)
```

### 4. Combining Multiple Logical Operators

```python
has_ticket = True
is_vip = False
is_banned = False

# The user can enter if they have a ticket OR are a VIP, AND they are NOT banned
can_enter = (has_ticket or is_vip) and (not is_banned)

print("Access granted?", can_enter)
```

# ⚡ Python Assignment Operators

Assignment operators assign or update the values stored inside a variable container. Compound shortcuts combine a math operation with an assignment in one step.

| Operator | Name | Example | Equivalent To |
| :---: | :--- | :---: | :--- |
| `=` | Assign | `x = 5` | `x = 5` |
| `+=` | Add and Assign | `x += 3` | `x = x + 3` |
| `-=` | Subtract and Assign | `x -= 2` | `x = x - 2` |
| `*=` | Multiply and Assign | `x *= 4` | `x = x * 4` |
| `/=` | Divide and Assign | `x /= 2` | `x = x / 2` |
| `%=` | Modulus and Assign | `x %= 3` | `x = x % 3` |
| `//=`| Floor Divide & Assign| `x //= 2`| `x = x // 2`|
| `**=`| Exponent and Assign | `x **= 2`| `x = x ** 2`|

### 1. Basic & Additive Assignment (=, +=)

```python
# Basic assignment
score = 10
print("Initial score:", score)

# Instead of typing: score = score + 5
score += 5  
print("After += 5:", score)
```

### 2. Subtraction & Multiplication Shortcuts (-=, *=)

```python
balance = 100

# Subtract 20 from balance
balance -= 20
print("After subtraction (-= 20):", balance)

# Double the remaining balance
balance *= 2
print("After multiplication (*= 2):", balance)
```

### 3. Division & Floor Division Shortcuts (/=, //=)

```python
items = 15

# Standard division shortcut (always converts the variable type to a float!)
items /= 2
print("After division (/= 2):", items, type(items))

# Resetting items to a whole number for floor division
count = 15
count //= 2
print("After floor division (//= 2):", count, type(count))
```

# 🆔 Python Identity Operators

Identity operators compare the memory locations (`id()`) of two objects to determine if they are actually the exact same entity.

| Operator | Description | Rule |
| :---: | :--- | :--- |
| `is` | Identity check | Returns `True` if both variables point to the **exact same memory address**. |
| `is not` | Negative identity check | Returns `True` if both variables point to **different memory addresses**. |

### ⚠️ `==` vs `is`
* `==` checks for **Equality** (Do the containers hold the same values?)
* `is` checks for **Identity** (Are they pointing to the exact same box in memory?)

### 1. True Identity sharing (Lists vs Variables)

```python
list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a  # list_c is given a direct link/reference to list_a

# Testing Equality (==) vs Identity (is)
print("Does list_a == list_b (Same values)?", list_a == list_b)
print("Does list_a is list_b (Same memory location)?", list_a is list_b)
print("Does list_a is list_c (Same memory location)?", list_a is list_c)
```

### 2. Checking Memory Locations manually (id())

```python
print("Memory address of list_a:", id(list_a))
print("Memory address of list_b:", id(list_b))
print("Memory address of list_c:", id(list_c))
```

### 3. The is not Operator

```python
x = ["apple", "banana"]
y = ["apple", "banana"]

# Since they are separate objects in memory, 'is not' evaluates to True
print("Are x and y different memory objects?", x is not y)
```

# 🔍 Python Membership Operators

Membership operators test for the presence of a specific element inside a collection or sequence (like strings, lists, tuples, or dictionaries).

| Operator | Description | Rule |
| :---: | :--- | :--- |
| `in` | Membership presence check | Returns `True` if the specified value is **found** inside the sequence. |
| `not in` | Membership absence check | Returns `True` if the specified value is **not found** inside the sequence. |

### 1. Checking Membership in a List (in)

```python
registered_users = ["alex", "emma", "david", "sarah"]

# Checking if a specific user exists in our data list
print("Is emma in the user list?", "emma" in registered_users)
print("Is james in the user list?", "james" in registered_users)
```

### 2. Checking Absence in a List (not in)

```python
banned_countries = ["Country A", "Country B"]
user_location = "Country C"

# Returns True because "Country C" is successfully absent from the ban list
can_join = user_location not in banned_countries
print("Is the user allowed to join?", can_join)
```

### 3. Searching Inside a String (Substrings)

```python
log_message = "CRITICAL ERROR: Database connection failed on port 5432"

# Checking for keywords inside text strings
print("Is 'ERROR' in the log?", "ERROR" in log_message)
print("Is 'SUCCESS' in the log?", "SUCCESS" in log_message)
```

### 4. Searching Inside a Dictionary

```python
product_stock = {"Apples": 50, "Bananas": 12, "Oranges": 0}

# Looks at keys directly
print("Do we sell Apples?", "Apples" in product_stock)

# This returns False because 12 is a value, not a dictionary key!
print("Is 12 a key in product stock?", 12 in product_stock)
```

# 🧠 Understanding Boolean Evaluation (`bool()`)

In Python, the built-in `bool()` function evaluates any value down to either `True` or `False`. Python divides all data into two logical categories:

1. **Falsy Values:** Values that represent emptiness, zero, or a null state always evaluate to `False`. For numbers, this means **exactly zero** (`0`, `0.0`, `0j`).
2. **Truthy Values:** Any numeric value that is **not exactly zero** (including negative numbers and complex numbers) always evaluates to `True`.


```python
val1 = bool(0)        # Integer zero (Falsy)
val2 = bool(3.14159)  # Non-zero float (Truthy)
val3 = bool(-3)       # Non-zero negative integer (Truthy)
val4 = bool(1.0+1j)   # Non-zero complex number (Truthy)


print(val1, val2, val3, val4)
```
