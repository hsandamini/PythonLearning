# 📥 Outputs and Inputs in Python

Python uses two simple, built-in functions to handle basic communication:
1. **`print()` (Output):** Sends data to the screen so the user can see it. It can display strings, numbers, variables, or expressions.
2. **`input()` (Input):** Pauses execution and waits for the user to type text on their keyboard. **Crucial Rule:** The `input()` function *always* saves user entry as a String (`str`), no matter what they type!

### 1. Printing a Simple String

```python
# Using print to print out string
print("This string is output to the screen!")
```

### 2. Combining a String and a Variable

```python
#Adding string and variable to print()
x = 10
print("The value of x is", x)
```

### 3. Separators and Ends

```python
# 1. Standard multiple item printing
print("Python", "Data", "Science")

# 2. Using custom separators (sep)
print("01", "02", "2026", sep="-") # Prints a clean date structure

# 3. Changing the line ending (end) so the next print statement stays on the same line
print("Hello ", end="")
print("World!")
```

### 🔧 Understanding Output Parameters: `sep` and `end`

The `print()` function has hidden default behaviors that we can manually override:
* **`sep` (Separator):** By default, Python places a blank space between multiple items. Changing `sep="*" ` overrides this space with an asterisk or any character of choice.
* **`end` (Line Ending):** By default, Python moves to a brand new line after executing a print statement. Changing `end="&"` overrides that behavior, replacing the newline break with a special character or preventing a line jump.

```python
print(1, 3, 5, 7, 8)
print(1, 3, 5, 7, 8, sep="*")
print(1, 3, 5, 7, 8, sep="#", end="&")
```

### 4. f-Strings

```python
user_name = "Alex"
lessons_completed = 15

# Add an 'f' right before the opening quote, and put variables inside curly braces {}
message = f"Welcome back {user_name}! You have completed {lessons_completed} lessons."
print(message)
```

### 5. Capturing User Input

```python
# The text inside the parentheses is the message prompt displayed to the user
favorite_food = input("Enter your favorite food: ")

print(f"Awesome! I love eating {favorite_food} too.")
```

### 6. The Input Number Trap

```python
# WRONG WAY: This will treat age as text, not a math number
age = input("Enter your age: ")
print(age + 1) # This would throw a TypeError!
```

```python
# CORRECT WAY: Instantly convert the typed text into a real integer number
user_age = int(input("Enter your age: "))
next_year_age = user_age + 1

print(f"Next year, you will be {next_year_age} years old!")
```

# 🎨 Formatting the Output in Python

Python provides three distinct techniques to inject variable data directly into text layouts:
1. **f-Strings (Formatted String Literals):** The modern, preferred standard introduced in Python 3.6.
2. **`str.format()` Method:** The older approach using curly braces `{}` as positional placeholders.
3. **`%` Operator (printf-style):** The legacy formatting style inherited from older computer languages.

### 1. f-Strings

```python
name = "Alice"
score = 98.5

# f-Strings make complex lines incredibly clean to read
announcement = f"Congratulations {name}! Your final data analysis score is {score}%."
print(announcement)
```

### 2. Advanced Float Rounding with f-Strings

```python
pi_value = 3.1415926535

# ': .2f' instructs Python to display only 2 digits past the decimal point
print(f"Pi rounded to two places: {pi_value:.2f}")
print(f"Pi rounded to three places: {pi_value:.3f}")
```

### 3. The Position Placeholder Style (.format())

```python
item = "Laptop"
price = 899

# Placeholders are filled sequentially by the values inside format()
receipt = "The price of the {} is ${}.".format(item, price)
print(receipt)
```

### 4. % Operator

```python
username = "admin"
attempts = 3

# %s stands for string placeholder; %d stands for integer placeholder
security_log = "User %s has made %d failed login attempts." % (username, attempts)
print(security_log)
```

# 📦 Importing Modules in Python

A **module** is a file containing Python statements and definitions. To use the code inside a module, we bring it into our script using the **`import`** keyword.

Python provides three primary techniques to import code components:
1. **`import module_name`:** Imports the entire module. You access components using dot notation (`module.function`).
2. **`from module_name import component`:** Imports a specific item directly, avoiding the need for dot notation.
3. **`import module_name as alias`:** Renames the module to a shorter nickname for cleaner code layout.

### 1. Standard Module Import (import)

```python
# Import the entire math library
import math

# Access constants and functions using dot notation (math.)
print("Value of PI:", math.pi)
print("Square root of 64:", math.sqrt(64))
```

### 2. Specific Component Import (from ... import)

```python
# Import only the 'pow' (power) and 'factorial' utilities from math
from math import pow, factorial

# Now we can use them directly without typing 'math.' out front
print("2 to the power of 5:", pow(2, 5))
print("Factorial of 5 (5!):", factorial(5))
```

### 3. Renaming with an Alias (import ... as)

```python
# Import the math module under the nickname 'm'
import math as m

# Call functions using your custom shorthand alias
print("Cosine of 0:", m.cos(0))
```

### 4. Exploring Module Contents (dir() function

```python
import math

# View the full index list of tools built into the math module
print(dir(math))
```
