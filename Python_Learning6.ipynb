# 🔀 Conditional Operations in Python

Conditional structures allow our programs to select different code blocks to run based on logical criteria. Every branch evaluates a statement down to a Boolean truth value.

* **`if`:** Runs a code block *only* if the test condition is `True`.
* **`if else`:** Adds an alternative route if the initial `if` test turns up `False`.
* **`if elif and else`:** Checks an ordered chain of rules sequentially before falling back.
* **`nested if`:** Places verification tests completely inside other successful decision pathways.

### 1. The if Statement

```python
# 1. Standalone conditional check
number = 10

if number > 5:
    print("Condition met: The number is greater than 5.")
```

### 2. The if else Statement

```python
# 2. Dual-branch pathway verification
age = 16

if age >= 18:
    print("Access approved: You are an adult.")
else:
    print("Access modified: You are a minor.")
```

### 3. The if elif and else Chain

```python
# 3. Multi-tier evaluation tree
score = 85

if score >= 90:
    print("Result: Distinction")
elif score >= 75:
    print("Result: First Class")
elif score >= 50:
    print("Result: Pass")
else:
    print("Result: Fail")
```

### 4. The nested if Statement

```python
# 4. Deep logical layer evaluation
has_id_card = True
is_employee = False

if has_id_card == True:
    print("Security Check 1: ID Card verified.")
    
    # Inner nested conditional layer
    if is_employee == True:
        print("Security Check 2: Welcome to the staff lounge.")
    else:
        print("Security Check 2: Access denied. Staff membership required.")
else:
    print("Security Check 1: Access denied. Missing ID card.")
```
