# 📦 Python Variables

A **variable** is a container for storing data values. 
* Python has **no command for declaring a variable**; it is created the moment you first assign a value to it using the `=` operator.
* Python variables use **Dynamic Typing**, meaning they do not need to be declared with any specific data type and can even change type after they have been set.

### 1. Basic Variable Assignment

```python
# Assigning values to variables
age = 25
first_name = "Sarah"

# Accessing and printing the variables
print(first_name)
print(age)
```

### 2. Changing (Overwriting) a Variable Value

```python
# A variable can change its value over time
score = 0
print("Initial score:", score)

score = 10  # The old value (0) is overwritten in memory
print("Updated score:", score)
```

### 3. Changing Data Types

```python
# Changing data types on the fly
x = 10          # x is currently an integer (number)
print(x)

x = "Hello"     # x is now a string (text)
print(x)
```

### 4. Multiple Assignment

```python
# Assigning multiple values to multiple variables
a, b, c = 1, 2, 3
print(a, b, c)

# Assigning the same value to multiple variables at once
x = y = z = "Python"
print(x, y, z)
```

## 📋 Official Rules for Valid Variable Names

To create a valid variable name in Python, you must follow these four strict criteria:
1. **Allowed Characters:** Can only contain alphanumeric characters (`A-z`, `0-9`) and underscores (`_`).
2. **Starting Rule:** Must begin with a letter or an underscore. It **cannot** start with a number.
3. **No Keywords:** Cannot match any of Python's reserved keywords (like `if`, `True`, or `pass`).
4. **No Spaces or Symbols:** Cannot contain spaces or special characters like `-`, `@`, `$`, or `%`.

```python
#  Correct Examples
user_age = 25          # Clear snake_case (Best practice)
_system_status = "OK"  # Allowed to start with an underscore
total2 = 150           # Numbers are allowed at the end or middle

print(user_age, _system_status, total2)
```

```python
# Wrong Examples
1variable = 10     <- Throws SyntaxError (Starts with a number)
user-age = 25      <- Throws SyntaxError (Hyphens/symbols not allowed)
global = True      <- Throws SyntaxError (Uses a reserved keyword)

print(1variable, user-age, global)
```

## ⚠️ Case Sensitivity and Advanced Variables

* Python is strictly **case-sensitive**. Changing a single letter's casing creates a completely new variable.
* Variables do not store data directly; they store a **reference link** to an object's location in your computer's memory.

### 1. Demonstrating Case Sensitivity

```python
# These are three completely distinct variables
age = 20
Age = 35
AGE = 50

print("Lowercase age:", age)
print("Capitalized Age:", Age)
print("Uppercase AGE:", AGE)
```

### 2. Checking Variable Memory Locations

```python
x = 100
y = 100

# Since both hold 100, Python optimizes memory by pointing both variables to the exact same box!
print("Memory address of x:", id(x))
print("Memory address of y:", id(y))

# If we change y, its memory address will shift to a new box
y = 200
print("New memory address of y:", id(y))
```

# 🔒 Python Constants

A **constant** is a variable whose value should remain unchanged throughout a program. 

* Python does not have built-in language support to block changing a variable's value.
* Instead, Python developers use **UPPERCASE** letters with underscores to flag a variable as a constant.
* Constants are typically grouped together at the very top of a Python file or kept in a separate file (e.g., `config.py`).

### 1. Standard Constants

```python
#  CORRECT CONVENTION: Use strictly uppercase letters
PI = 3.14159
GRAVITY = 9.8
MAX_LOGIN_ATTEMPTS = 5

print("The value of PI is:", PI)
print("The value of GRAVITY is:", GRAVITY)
print("Maximum login attempts allowed:", MAX_LOGIN_ATTEMPTS)
```

### 2. The Python Flexibility in Constants

```python
# Even though it is named like a constant, Python will technically let you overwrite it
TOTAL_MONTHS = 12
print("Original:", TOTAL_MONTHS)

# This runs without error, but it breaks the constant rule!
TOTAL_MONTHS = 13 
print("Accidentally modified:", TOTAL_MONTHS)
```

# 🔢 Python Numeric Literals

A **numeric literal** is a specific numeric value written directly into your code. Python splits numeric literals into three clear data types based on how they are written:

1. **Integer (int):** Whole numbers without any decimal point (can be positive or negative).
2. **Float (float):** Numbers containing a decimal point or an exponential notation (`e`).
3. **Complex (complex):** Numbers written in the format `real + imaginary`, where `j` represents the imaginary unit.
    

## 🔢 Advanced Numeric Literal Systems

Python allows integers to be defined using four separate number systems by adding a specific prefix:
* **Binary (Base 2):** Prefix code `0b` or `0B`
* **Octal (Base 8):** Prefix code `0o` or `0O`
* **Decimal (Base 10):** No prefix (standard numbers)
* **Hexadecimal (Base 16):** Prefix code `0x` or `0X`

When you print these variables, Python automatically converts them into standard decimal format in the output block.

### 1. Integer Literals (Whole Numbers)

```python
# Positive, negative, and zero integer literals
items_count = 50
temperature = -5
zero_value = 0

print(items_count, type(items_count))
print(temperature, type(temperature))
```

### 2. Floating-Point Literals (Decimals)

```python
# Standard decimal literals
price = 19.99
negative_float = -0.0035

# Scientific/Exponential notation (e3 means multiplied by 10 to the power of 3)
distance = 1.5e3  # Evaluates to 1500.0

print(price, type(price))
print(distance, type(distance))
```

### 3. Complex Literals

```python
# Complex number literals (Real + Imaginary)
coordinates = 3 + 5j

print(coordinates, type(coordinates))
print("Real part:", coordinates.real)
print("Imaginary part:", coordinates.imag)
```

### 4. The Underscore Trick for Large Numbers 

```python
# Clean formatting for large numbers
one_million = 1_000_000
gas_constant = 8.314_462

print(one_million)    # Prints as a normal integer: 1000000
print(gas_constant)   # Prints as a normal float: 8.314462
```

### 5.Number System Prefixes

```python
# Non-decimal integer literal bases
binary_num = 0b1010
octal_num = 0o310
hex_num = 0x12c

print(binary_num, type(binary_num))
print(octal_num, type(octal_num))
print(hex_num, type(hex_num))
```

# 🔤 Python String Literals

A **string literal** is text wrapped in matching quote characters. Python treats all text data inside quotes as part of the `str` class. 
* Single and double quotes work identically for single lines.
* Triple quotes are used to handle multi-line blocks of text.

### 1.Single and Double Quotes

```python
# Creating strings using single and double quotes
message1 = "Hello, Python!"
message2 = 'Learning data science'

print(message1, type(message1))
print(message2, type(message2))
```

### 2.Multi-line Strings

```python
# Triple quotes allow text to span across multiple lines seamlessly
multiline_text = """This is a string literal
that spans across multiple
different lines of text."""

print(multiline_text, type(multiline_text))
```

### 3.Handling Quotes Inside Strings

```python
# Use double quotes if you need an apostrophe inside your text
sentence1 = "It's a beautiful day to code!"

# Use single quotes if you need double quotes inside your text
sentence2 = 'The instructor said, "Practice every day!"'

print(sentence1)
print(sentence2)
```

### 4.Character Literals

```python
# Many languages have a specific 'char' type for single letters.
# Python does NOT have a char type! A single letter is just a string of length 1.
letter = "A"

print(letter, type(letter))
```

# ⚖️ Python Boolean Literals

A **Boolean literal** represents a binary truth value. 

* There are only two built-in Boolean values: `True` and `False`.
* They belong to the `<class 'bool'>` data type.
* **The Numeric Value Rule:** Behind the scenes, Python treats `True` as equal to `1` and `False` as equal to `0` (`True == 1` and `False == 0`). This means you can actually use them in mathematical equations!
* **Crucial Case-Sensitivity Rule:** They must start with capital letters. Writing `true` or `false` in lowercase will cause a `NameError`.

```python
# True equals 1 and False equals 0
print("Does True == 1?", True == 1)
print("Does False == 0?", False == 0)

# We can even add them together!
calculation1 = True + True   # 1 + 1
calculation2 = True + False  # 1 + 0

print("True + True =", calculation1)
print("True + False =", calculation2)
```

### 1. Assigning Boolean Literals

```python
# Assigning boolean values directly to variables
is_python_fun = True
is_script_broken = False

print(is_python_fun, type(is_python_fun))
print(is_script_broken, type(is_script_broken))
```

### 2. Creating Booleans via Comparisons

```python
# Boolean values are automatically created when you compare data
result1 = (10 > 5)   # Evaluates to True
result2 = (3 == 9)   # Evaluates to False

print("Is 10 greater than 5?", result1)
print("Is 3 equal to 9?", result2)
```

### 3. The Case-Sensitivity Error Case

```python
# WRONG: Lowercase true/false will crash the interpreter
active_status = true
```

# 🕳️ Python Special Literals

Python has exactly one special literal: **`None`**.

* It is used to signify the **absence of a value** or a null state.
* It belongs to its own unique data type: `<class 'NoneType'>`.
* Like Booleans, it is strictly case-sensitive and must start with a capital letter (**`None`**).
* It is commonly used as a placeholder for variables that will receive data later.

### 1. Declaring and Testing None

```python
# Setting up an empty variable placeholder
user_profile_picture = None

print(user_profile_picture)
print(type(user_profile_picture))
```

### 2. Using None in a Practical Decision (If Statement)

```python
# Simulating a system checking if data has been filled out yet
current_user = None

if current_user is None:
    print("No user is logged in. Redirecting to login page...")
else:
    print("Welcome back!")
```

### 3. Functions that Return Nothing

```python
# A simple function that just prints text
def greet_user():
    print("Hello!")

# Capturing the return value of that function
result = greet_user()

print("The return value of the function is:", result)
```

# 🗃️ Python Collection Literals

Collection literals allow you to group multiple data values inside a single variable container. Python features four distinct built-in collections:

1. **List `[]`:** An ordered, mutable (changeable) collection that allows duplicate items.
2. **Tuple `()`:** An ordered, immutable (unchangeable) collection that allows duplicate items.
3. **Dictionary `{key: value}`:** An unordered collection of key-value pairs. Keys must be unique.
4. **Set `{}`:** An unordered, unindexed collection of unique items. No duplicates allowed.

### 1. List Literals (Using Square Brackets [])

```python
# A list can hold mixed data types and allows duplicates
shopping_list = ["apple", "banana", "apple", 100, 5.5]

print(shopping_list)
print(type(shopping_list))
```

### 2. Tuple Literals (Using Parentheses ())

```python
# A tuple is a protected, unchangeable collection
gps_coordinates = (40.7128, -74.0060)

print(gps_coordinates)
print(type(gps_coordinates))
```

### 3. Dictionary Literals (Using Curly Braces with Colons {key: value})

```python
# Mapping data using key-value properties
user_account = {
    "username": "coder_99",
    "years_active": 3,
    "is_premium": True
}

print(user_account)
print(type(user_account))
print("The username is:", user_account["username"])
```

### 4. Set Literals (Using Curly Braces {})

```python
# Sets automatically remove duplicate entries completely
unique_numbers = {1, 2, 2, 3, 4, 4, 4, 5}

print(unique_numbers)  # Notice how duplicate numbers are gone in the output!
print(type(unique_numbers))
```

### Examples with all literal collections

```python
# Initializing different collection types
fruits = ["fig", "lemon", "banana"]                    # List
numbers = (1, 2, 3, 4)                                 # Tuple
words = {"first word": "Banana", "Second word": "Hi"}  # Dictionary
chars = {'A', 'B', 'C'}                                # Set
```

```python
# Printing out the collection contents
print("List content:", fruits)
print("Tuple content:", numbers)
print("Dict content:", words)
print("Set content:", chars)

print("-" * 40)

# Verifying the built-in python classes
print(type(fruits))
print(type(numbers))
print(type(words))
print(type(chars))
```
